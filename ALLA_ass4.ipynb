```python
import requests
import folium
from datetime import datetime
import time
import threading


```

```python
start_time = datetime.now()
start_time
```

```python
lat=0
long=0
def data1():
    
    json_file = "http://api.open-notify.org/iss-now.json"
    y = 3
    x = 0

    
    while x < y:
        response = requests.get(json_file)
        response.raise_for_status()
        data = response.json()
        time_stamp = data['timestamp']
        print(time_stamp)
        longitude = float(data['iss_position']['longitude'])
        latitude = float(data['iss_position']['latitude'])
        x=x+1
        lat=latitude
        long=longitude
    return time_stamp, longitude, latitude

```

```python
# Function to update the map with markers
def update_map():
    while not exit_event.is_set():
        info = data1()
        time_stamp, longitude, latitude = info
        marker = folium.Marker([latitude, longitude], icon=None)
        marker.add_to(map)
        time.sleep(5)
```

```python
map = folium.Map(location=[lat, long], zoom_start=2)
map
```

```python
exit_event = threading.Event()

# Start the update_map thread
update_map_thread = threading.Thread(target=update_map)
update_map_thread.start()


time.sleep(3600)

exit_event.set()
```

```python
update_map_thread.join()
```
